Stack plot

```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]
plt.plot([],[], color='m', label='Sleeping')
plt.plot([],[], color='c', label='Eating')
plt.plot([],[], color='r', label='Working')
plt.plot([],[], color='k', label='Playing')
plt.stackplot(days, sleeping, eating, working ,playing, colors=['m','c', 'r', 'k'])
plt.xlabel('days')
plt.ylabel('Activities')
plt.title('Persons Weekly Spent Time per activities\n Spring 2018')
plt.legend()
plt.show()
```

Pie Chart

```python
slices = [39,14,26,41]
activities = ['sleeping', 'eating', 'working','playing']
cols = ['c','m','r', 'b']
plt.pie(slices,labels= activities,colors= cols,startangle=100,shadow=True,explode = (0,0,0.2,0),autopct = '%1.1f%%')
plt.title('Persons Weekly Spent Time per activities\n Spring 2018')
plt.legend()
plt.show()
```

```python
# Simple stripplot
import seaborn as sns
sns.stripplot( x =dataset['salary'])
```

```python
# Stripplot over categories
sns.stripplot( x = dataset['rank'], y= dataset['salary'],data=dataset);
```

```python
# Stripplot over categories
sns.stripplot( x = dataset['discipline'], y =dataset['salary'],data=dataset ,  jitter=1)
#Jitter is adding a small amount of variability (horizontal or vertical) to the data to ensure all data points are visible.
```

```python
sns.stripplot( x= dataset['salary'], y = dataset['discipline'],data=dataset, jitter=True);
```

```python
sns.stripplot( x = dataset['rank'], y= dataset['salary'],data=dataset, jitter=True);
```

```python
# Add hue to the graph
# Stripplot over categories
sns.stripplot( x ='sex', y= 'salary', hue='rank',data=dataset, jitter=True )
```

```python
# Draw data on top of boxplot
import numpy as np
sns.boxplot(x = 'salary', y ='sex', data=dataset, whis=np.inf )
sns.stripplot(x = 'salary', y ='sex', data=dataset,jitter=True, color='0.02')
```
