```python
#Creating a data frame including NaN
import pandas as pd
import numpy as np
dataset = pd.DataFrame(np.random.randn(5, 3),
index=['a', 'c', 'e', 'f', 'h'],columns=['one',
'two', 'three'])
dataset.rename(columns={"one":'stock1',"two":'stock2',"three":'stock3'}, inplace=True)
dataset = dataset.reindex(['a', 'b', 'c', 'd', 'e',
'f', 'g', 'h'])
print (dataset)
```

```python
print(dataset.isnull())
```

```python
#Replacing NaN with a Scalar Value
print (dataset)
dataset.fillna(0)
```

```python
# Fill missing values forward
print (dataset)
dataset.fillna(method='bfill')
```

```python
# Fill missing values backward
print (dataset)
dataset.fillna(method='bfill')
```

```python
print (dataset)
dataset.dropna()
```

```python
#Using the replace() Function
dataset.replace(np.nan, 0 )
```

```python
#Using the replace() Function
print (dataset)
 
```

```python
print(dataset['stock1'].mean())
print(dataset['stock2'].mean())
print(dataset['stock3'].mean())
dataset[['stock1','stock2','stock3']].fillna(value=dataset[['stock1','stock2','stock3']].mean())
```
