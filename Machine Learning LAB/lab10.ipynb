```python
from sklearn.datasets import load_iris
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
import pandas as pd
```

```python
iris=load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
```

```python
x=iris.data[:,2:4]
y=iris.target
svm=SVC(kernel='rbf',gamma=0.5,C=1.0)
svm.fit(x,y)

DecisionBoundaryDisplay.from_estimator(svm,x,response_method='predict',cmap=plt.cm.Spectral,aplha=0.8,xlabel=iris.feature_names[0],ylabel=iris.feature_names[1],)

plt.scatter(x[:,0],x[:,1],c=y,s=20,edgecolors='k')
plt.show()

svm=SVC(kernel='poly',degree=3,C=1.0)
svm.fit(x,y)

DecisionBoundaryDisplay.from_estimator(svm,x,response_method='predict',cmap=plt.cm.Spectral,aplha=0.8,xlabel=iris.feature_names[0],ylabel=iris.feature_names[1],)

plt.scatter(x[:,0],x[:,1],c=y,s=20,edgecolors='k')
plt.show()

svm=SVC(kernel='linear',C=1.0)
svm.fit(x,y)

DecisionBoundaryDisplay.from_estimator(svm,x,response_method='predict',cmap=plt.cm.Spectral,aplha=0.8,xlabel=iris.feature_names[0],ylabel=iris.feature_names[1],)

plt.scatter(x[:,0],x[:,1],c=y,s=20,edgecolors='k')
plt.show()

svm=SVC(kernel='sigmoid',C=1.0)
svm.fit(x,y)

DecisionBoundaryDisplay.from_estimator(svm,x,response_method='predict',cmap=plt.cm.Spectral,aplha=0.8,xlabel=iris.feature_names[0],ylabel=iris.feature_names[1],)

plt.scatter(x[:,0],x[:,1],c=y,s=20,edgecolors='k')
plt.show()
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.inspection import DecisionBoundaryDisplay

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, 2:4]  # Using only two features for visualization
y = iris.target

# Create a dictionary to store kernel names and their respective SVC models
kernel_models = {
    'Linear': SVC(kernel='linear', C=1),
    'Polynomial': SVC(kernel='poly', degree=3, C=1),
    'RBF': SVC(kernel='rbf', gamma=0.5, C=1),
    'Sigmoid': SVC(kernel='sigmoid', C=1),
}

# Create subplots to display decision boundaries for each kernel
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.suptitle('Decision Boundaries with Different Kernel Functions')

for (kernel_name, svc), ax in zip(kernel_models.items(), axes.ravel()):
    svc.fit(X, y)
    DecisionBoundaryDisplay.from_estimator(
        svc, X, response_method='predict', cmap=plt.cm.Spectral, ax=ax
    )
    ax.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors='k')
    ax.set_title(f'Kernel: {kernel_name}')

plt.show()
```
