```python
def estimate_coef(x,y):
    n=np.size(x)
    x_bar=np.mean(x)
    y_bar=np.mean(y)
    nr=np.sum(x*y)-(n*x_bar*y_bar)
    dr=np.sum(x*x)-(n*x_bar*x_bar)
    m=nr/dr
    c=y_bar-m*x_bar
    return(m,c)

def plot_regression_line(x,y,b):
    plt.scatter(x,y,color='r',marker='o',s=30)
    y_pred=b[0]*x+b[1]
    e=np.mean(y_pred-y)
    plt.plot(x,y_pred,color='b',linewidth=3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    return e

def main():
    x=df['Year'].to_numpy()
    y=df['Price'].to_numpy()
    b=estimate_coef(x,y)
    print("Slope= ",b[0])
    print("Intercept",b[1])
    print("linear regression eqn={}x+{}".format(b[0],b[1]))
    p=plot_regression_line(x,y,b)
    print("prediction for year 2025")
    y_pred=b[0]*2025+b[1]
    print(y_pred)
    print(f"error={p}")
    

if __name__ == "__main__":
    main()
    
```

```python
#USING LIBRARIES
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data=df
X=data['Year'].to_numpy()
X=X.reshape(-1,1)
Y=data['Price']
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.3, random_state = 10 )  

regressor=LinearRegression()
regressor.fit(X_train,Y_train)
y_pred= regressor.predict(X_test)
print(y_pred)
r_sq=regressor.score(X,Y)
print(f"Coefficient of determination={r_sq}")
print(f"intercept: {regressor.intercept_}")
print(f"slope: {regressor.coef_}")
print(f"predicted response:\n{y_pred}")
```

```python
year=[2025]
year=np.array(year).reshape(-1,1)
pred=regressor.predict(year)
print(pred)
```
