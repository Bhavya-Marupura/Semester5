```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salary=[1.7,2.4,2.3,3.1,3.7,4.2,4.4,6.1,5.4,5.7,6.4,6.2]
exp=[1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1,3.3,3.7,4.2,4.4]
df=pd.DataFrame(data={'Salary':salary,'Experience':exp})
y=df['Salary'].to_numpy()
x=df['Experience'].to_numpy()
b0=0
b1=0
alpha=0.01
beta=[]
error=[]
iterations=[]
itr=1
for i in range(5):
    for j in range(12):
        iterations.append(itr)
        y_pred=(b1*x[j])+b0
        e=y_pred-y[j]
        error.append(e)
        b0=b0-(alpha*e*1)
        b1=b1-(alpha*e*x[j])
        beta.append(b1)
        itr=itr+1
        
plt.plot(iterations,error,color='b')
plt.show()
plt.plot(beta,error,color='b')
plt.show()
print(f"b1={b1}")
print(f"b0={b0}")
```

```python
from sklearn.linear_model import SGDRegressor
x=x.reshape(-1,1)
reg=SGDRegressor(max_iter=60,alpha=alpha,random_state=0)
model=reg.fit(x,y)
y_pred=model.predict(x)
print(f"b1={model.coef_}")
print(f"b0={model.intercept_}")
```
