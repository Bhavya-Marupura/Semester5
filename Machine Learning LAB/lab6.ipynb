```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data={'Salary':[1.7,2.4,2.3,3.1,3.7,4.2,4.4,6.1,5.4,5.7,6.4,6.2],'Experience':[1.2,1.5,1.9,2.2,2.4,2.5,2.8,3.1,3.3,3.7,4.2,4.4]}
df=pd.DataFrame(data)
df
```

```python
x=df['Experience'].to_numpy()
y=df['Salary'].to_numpy()
n=np.size(x)
b=[0,0]
x_bar=np.mean(x)
y_bar=np.mean(y)
nr=np.sum(x*y)-(n*x_bar*y_bar)
dr=np.sum(x*x)-(n*x_bar*x_bar)
b[0]=nr/dr
b[1]=y_bar-b[0]*x_bar
print("Slope= ",b[0])
print("Intercept",b[1])
print("linear regression eqn={}x+{}".format(b[0],b[1]))
y_pred=b[0]*x+b[1]
plt.scatter(x,y,color='r',marker='o',s=20,label='Original dataset values')
plt.plot(x,y_pred,color='b',linewidth=3, label='Predicted dataset values')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

e=np.mean((y-y_pred)*(y-y_pred))
print(f"MSE:{e}")
```

```python
#a
slope=[0.1,1.5,0.8]
intercept=1.1
pred=[]
for i in range(len(slope)):
    y_pred=slope[i]*x+intercept
    pred.append(y_pred)
pred=np.array(pred)
print('Predicted values for slopes 0.1,1.5,0.8')
print(pred)
mse=[]
for i in range(len(pred)):
    e=np.mean((pred[i]-y)*(pred[i]-y))
    mse.append(e)
print("\n")
print(f"MSE:{mse}")
#GRAPH 1 WITH SLOPE 0.1
plt.scatter(x,y,color='r',marker='o',s=20,label='Original dataset values')
plt.plot(x,pred[0],color='b',linewidth=3, label='Predicted dataset values')
plt.xlabel('Experience')
plt.title('Slope 0.1, Intercept 1.1')
plt.ylabel('Salary')
plt.legend()
plt.show()
print(f"MSE for graph1={mse[0]}")

#GRAPH 2 WITH SLOPE 1.5
plt.scatter(x,y,color='r',marker='o',s=20,label='Original dataset values')
plt.plot(x,pred[1],color='b',linewidth=3, label='Predicted dataset values')
plt.xlabel('Experience')
plt.title('Slope 1.5, Intercept 1.1')
plt.ylabel('Salary')
plt.legend()
plt.show()
print(f"MSE for graph2={mse[1]}")

#GRAPH 3 WITH SLOPE 0.8
plt.scatter(x,y,color='r',marker='o',s=20,label='Original dataset values')
plt.plot(x,pred[2],color='b',linewidth=3, label='Predicted dataset values')
plt.xlabel('Experience')
plt.title('Slope 0.8, Intercept 1.1')
plt.ylabel('Salary')
plt.legend()
plt.show()
print(f"MSE for graph3={mse[2]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

learning_rate = 0.01

np.random.seed(0)
B0 = np.random.rand()
B1 = np.random.rand()
error_values = []
iterations = []

total_iterations = 60
epochs = 5

for iteration in range(total_iterations):
    for _ in range(epochs):
        indices = np.random.permutation(len(x))
        X_shuffled = x[indices]
        y_shuffled = y[indices]
        
        for i in range(len(x)):
            y_pred = B0 + B1 * X_shuffled[i]
            
            error = y_pred - y_shuffled[i]
            
            B0 -= learning_rate * error
            B1 -= learning_rate * error * X_shuffled[i]
        
        mse = np.mean((y - (B0 + B1 * x)) ** 2)
        
        error_values.append(mse)
        iterations.append(iteration * epochs + _)

plt.plot(iterations, error_values)
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Error vs. Iteration ')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error

sgd_model = SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=60 * 5, random_state=0)

sgd_model.fit(x.reshape(-1, 1), y)

B0_sgd = sgd_model.intercept_[0]
B1_sgd = sgd_model.coef_[0]

y_pred_sgd = sgd_model.predict(x.reshape(-1, 1))
mse_sgd = mean_squared_error(y, y_pred_sgd)

print("Results using scikit-learn SGD:")
print("B0 (Intercept):", B0_sgd)
print("B1 (Slope):", B1_sgd)
print("MSE:", mse_sgd)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.plot(iterations, error_values)
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Error vs. Iteration (SGD from Scratch)')

plt.subplot(122)
beta_values_sgd = np.linspace(B1_sgd - 0.1, B1_sgd + 0.1, num=100)
error_values_sgd = []

for beta in beta_values_sgd:
    y_pred_sgd = B0_sgd + beta * x
    mse_sgd = mean_squared_error(y, y_pred_sgd)
    error_values_sgd.append(mse_sgd)

plt.plot(beta_values_sgd, error_values_sgd)
plt.xlabel('Beta (B1)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Error vs. Beta (SGD with scikit-learn)')
plt.tight_layout()
plt.show()
```
