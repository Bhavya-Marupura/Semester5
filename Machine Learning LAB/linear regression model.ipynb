```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('gold_rate.csv')
x=df['Year'].to_numpy()
x=x.reshape(-1,1)
y=df['Price'].to_numpy()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(f"Slope: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"coefficient of determination: {model.score(x,y)}")
print(f"predicted response= {y_pred}")
print(f"error={mean_squared_error(y_test,y_pred)}")
y_pr=model.predict([[2025]])
print(f"2025 gold rate: {y_pr}")
plt.scatter(x_train,y_train,color='r',marker='o',label='training set')
plt.plot(x_test,y_pred,color='b',linewidth=3, label='predicted data')
plt.title("gold rate pred model")
plt.xlabel("year")
plt.ylabel('gold rate')
plt.legend()
plt.show()
```

```python
data=pd.read_csv('gold_rate.csv')
x=data['Year'].to_numpy()
y=data['Price'].to_numpy()
b=[0,0]
x_bar=np.mean(x)
y_bar=np.mean(y)
n=np.size(x)
nr=np.sum(x*y)-(n*x_bar*y_bar)
dr=np.sum(x*x)-(n*x_bar*x_bar)
b[0]=nr/dr
b[1]=y_bar-(x_bar*b[0])
print(b[0])
print(b[1])
y_pred=b[0]*x+b[1]
plt.scatter(x,y,color='r',marker='o',label='training set')
plt.plot(x,y_pred,color='b',linewidth=3, label='predicted data')
plt.title("gold rate pred model")
plt.xlabel("year")
plt.ylabel('gold rate')
plt.legend()
plt.show()
print(f'2025 gold rate:{b[0]*2025+b[1]}')
print(f"error={np.mean((y-y_pred)*(y-y_pred))}")
```

```python
df=pd.read_csv('hepatitis_csv.csv')
df=df.fillna(df.median())
x=df.iloc[:,0].to_numpy()
x=x.reshape(-1,1)
y=df.iloc[:,-4].to_numpy()
print(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(f"Slope: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"coefficient of determination: {model.score(x,y)}")
print(f"predicted response= {y_pred}")

plt.scatter(x_train,y_train,color='r',marker='o',label='training set')
plt.plot(x_test,y_pred,color='b',linewidth=3, label='predicted data')
plt.yticks(np.arange(0,10,1))
plt.xticks(np.arange(0,100,10))
plt.title("hepatitis pred model")
plt.xlabel("age")
plt.ylabel('albumin')
plt.legend()
plt.show()
```
