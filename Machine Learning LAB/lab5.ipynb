```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
np.random.seed(0)
x=np.random.rand(100,1)
y=4+3*x+np.random.randn(100,1)
degree=2
poly_features=PolynomialFeatures(degree=degree)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly,y)
x_new=np.linspace(0,2,100).reshape(-1,1)
x_new_poly=poly_features.fit_transform(x_new)
y_new=model.predict(x_new_poly)
plt.scatter(x,y,label='original data')
plt.plot(x_new,y_new,'r-',label='poly regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
coefficients = model.coef_
intercept = model.intercept_
print("Coefficients:")
print(coefficients)
print("Intercept:")
print(intercept)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
df=pd.read_csv('gold_rate.csv')
x=df['Year'].to_numpy().reshape(-1,1)
y=df['Price'].to_numpy()
y=y/10
degree=5
poly_features=PolynomialFeatures(degree=degree)
x_poly=poly_features.fit_transform(x)
poly_features.fit(x_poly,y)
lin=LinearRegression()
lin.fit(x,y)

lin2=LinearRegression()
lin2.fit(x_poly,y)

plt.scatter(x, y, color='blue')
plt.plot(x, lin.predict(x), color='red')
plt.title('Linear Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.show()

plt.scatter(x,y,color='b',label='original data')
plt.plot(x,lin2.predict(poly_features.fit_transform(x)),color='r')
plt.title('Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
 
plt.show()

coefficients = lin2.coef_
intercept = lin2.intercept_
print("Coefficients:")
print(coefficients)
print("Intercept:")
print(intercept)
a=np.array([2025]).reshape(-1,1)
y_pred=lin2.predict(poly_features.fit_transform(a))
print(y_pred)
```
